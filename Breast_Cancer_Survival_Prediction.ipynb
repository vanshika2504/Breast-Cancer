## ***6. Feature Engineering & Data Pre-processing***

#### What all missing value imputation techniques have you used and why did you use those techniques?

We used two missing value imputation techniques:

1. **Imputing with Median Date**: For the 'Date_of_Last_Visit' column, we imputed missing values with the median date. We chose this technique because dates are continuous variables, and imputing with the median preserves the central tendency of the data without being influenced by outliers.

2. **Imputing with Most Frequent Value**: For the 'Patient_Status' column, we imputed missing values with the most frequent value (mode) in the column. We chose this technique because 'Patient_Status' is a categorical variable, and imputing with the mode allows us to fill missing values with the most common category, which is often a reasonable assumption for categorical data.

These techniques were chosen based on the nature of the data in each column and aimed to preserve the integrity of the dataset while handling missing values effectively.


### 3. Categorical Encoding

````python
# Convert 'Tumour_Stage' to numerical values
df['Tumour_Stage'] = df['Tumour_Stage'].map({'I': 1, 'II': 2, 'III': 3})

# Convert 'Histology' to numerical values
df['Histology'] = df['Histology'].map({'Infiltrating Ductal Carcinoma': 1, 'Infiltrating Lobular Carcinoma': 2, 'Mucinous Carcinoma': 3})

# Convert 'ER status' to numerical values
df['ER status'] = df['ER status'].map({'Positive': 1})

# Convert 'PR status' to numerical values
df['PR status'] = df['PR status'].map({'Positive': 1})

# Convert 'HER2 status' to numerical values
df['HER2 status'] = df['HER2 status'].map({'Positive': 1, 'Negative': 2})

# Convert 'Gender' to numerical values
df['Gender'] = df['Gender'].map({'MALE': 1, 'FEMALE': 2})

# Convert 'Surgery_type' to numerical values
df['Surgery_type'] = df['Surgery_type'].map({'Other': 1, 'Modified Radical Mastectomy': 2, 'Lumpectomy': 3, 'Simple Mastectomy': 4})

# Display the first few rows of the modified dfFrame
df.head()
````

#### What all categorical encoding techniques have you used & why did you use those techniques?

## Categorical Encoding Techniques Used

In the project, we utilized a combination of ordinal encoding and one-hot encoding techniques to convert categorical variables into numerical values.

### Ordinal Encoding

Ordinal encoding assigns numerical values to categories based on their order or rank. It is suitable for variables where there is a logical order among the categories. In our dataset, ordinal encoding was applied to variables such as 'Tumour_Stage', 'ER status', 'PR status', and 'HER2 status'.

### One-Hot Encoding

One-hot encoding creates binary columns for each category in the variable. It is suitable for nominal categorical variables where there is no inherent order among the categories. In our dataset, one-hot encoding was used for variables such as 'Surgery_type' and 'Histology' to avoid imposing any ordinal relationship among the categories.



### 7. Data Splitting

````python
x=df[["Age","Gender","Protein1","Protein2","Protein3","Protein4","Tumour_Stage","Histology","ER status","PR status","HER2 status","Surgery_type"]]
y=df["Patient_Status"]
from sklearn.model_selection import train_test_split

#Splitting data into train and test data
X_train, X_test, y_train, y_test=train_test_split(x,y, test_size=0.2,random_state=42)
````

We've chosen an 80:20 data splitting ratio, with 80% of the data allocated for training and 20% for testing. This split strikes a balance between providing enough data for the model to learn from and ensuring sufficient unseen data for evaluation. It helps prevent overfitting to the test set and allows for effective model evaluation and generalization to new data.


## ***7. ML Model Implementation***

### ML Model - 1

````python
# ML Model - 1 Implementation
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the Logistic Regression model
log_reg_model = LogisticRegression(max_iter=1000)

# Train the model
log_reg_model.fit(X_train, y_train)

# Predict on the test set
y_pred = log_reg_model.predict(X_test)

# Calculate the accuracy score
accuracy_lr = accuracy_score(y_test, y_pred)
precision_lr = precision_score(y_test, y_pred, average='weighted',zero_division=1)
recall_lr = recall_score(y_test, y_pred, average='weighted')
f1_lr = f1_score(y_test, y_pred, average='weighted')
print("Accuracy Score:", accuracy_lr)


# Predict on the model
# Input features for prediction
new_data = np.array([[36.0, 1, 0.080353, 0.42638, 0.54715, 0.273680, 3, 1, 1, 1, 2, 2]])

# Predict class on new data
predicted_class = log_reg_model.predict(new_data)
print("Predicted class:", predicted_class)
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
# Evaluation metric scores
evaluation_metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
scores = [accuracy_lr, precision_lr, recall_lr, f1_lr]

# Create bar plot
plt.figure(figsize=(8, 6))
plt.bar(evaluation_metrics, scores, color=['blue', 'green', 'orange', 'red'],ec='black')
plt.title('Evaluation Metric Scores (Logistic Regression)')
plt.xlabel('Metric')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.show()
````

### ML Model - 2

````python
# ML Model - 2 Implementation
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the Decision Tree model
decision_tree_model = DecisionTreeClassifier()

# Train the model
decision_tree_model.fit(X_train, y_train)

# Predict on the test set
y_pred = decision_tree_model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

# Print evaluation metrics
print("Accuracy Score:", accuracy)

# Predict on the model
new_data = np.array([[45.0, 2, 1.245, 0.972, 0.789, 0.654, 2, 2, 1, 2, 1, 3]])

# Predict class on new data
predicted_class = decision_tree_model.predict(new_data)
print("Predicted class:", predicted_class)
````
